```python
import os
import pandas
 
DATA = pandas.read_csv("Data.csv")

DATA

```

```python
# Separating the depenedent and independent data variables into two dataframes.
from sklearn.model_selection import train_test_split 
X = DATA.drop(['cnt'],axis=1) # use others as inputs for prediction
Y = DATA['cnt'] # set cnt as output for prediction

type(X)

Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.20, random_state=0)

X_train
```

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso(alpha = 1.0) # set penalty value
lasso = lasso_model.fit(X_train , Y_train)
lasso_predict = lasso.predict(X_test)
```

```python
import numpy as np
error = np.mean(np.abs((Y_test - lasso_predict)))
print('error is', error)
```

```python
# Mean absolute percentage error

import numpy as np
def MAPE(Y_actual,Y_Predicted):
    mape = np.mean(np.abs((Y_actual - Y_Predicted)/Y_actual))*100
    return mape
```

```python
Lasso_MAPE = MAPE(Y_test,lasso_predict)
print("MAPE value: ",Lasso_MAPE)
```

```python
Accuracy = 100 - Lasso_MAPE
print('Accuracy of Lasso Regression: {:0.2f}%.'.format(Accuracy))
```
